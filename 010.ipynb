# 010.图像的拼接(np.hstack)

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def show(img):
    cv2.namedWindow("Img",cv2.WINDOW_NORMAL)
    cv2.resizeWindow("Img",600,450)
    cv2.imshow("Img",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img01 = cv2.imread("./data/1_BF.jpg")
img02 = cv2.imread("./data/1_DAPI.jpg")

img01 = cv2.resize(img01,(400,400))
img02 = cv2.resize(img02,(400,400))

img03 = np.hstack((img01,img02))
img04 = np.vstack((img01,img02))

plt.subplot(211),plt.title("HSTACK"),plt.axis("off")
plt.imshow(img03)

plt.subplot(212),plt.title("VSTACK"),plt.axis("off")
plt.imshow(img04)

show(img04)
```

# 04.使用matplotlib显示图像


```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
img01 = cv2.imread("./dataplt/dapi.png")
img02 = cv2.imread("./dataplt/dapi2.png")
img03 = cv2.imread("./dataplt/dapi3.png")

img01 = cv2.cvtColor(img01,cv2.COLOR_BGR2RGB)
img02 = cv2.cvtColor(img02,cv2.COLOR_BGR2RGB)
img03 = cv2.cvtColor(img03,cv2.COLOR_BGR2RGB)

#plt.rcParams['font.sans-serif'] = ['yahei']
plt.subplot(131),plt.title("1. RGB"),plt.axis("off")
plt.imshow(img01)

plt.subplot(132),plt.title("2. RGB"),plt.axis("off")
plt.imshow(img02)

plt.subplot(133),plt.title("3. GRAY"),plt.axis("off")
plt.imshow(img03,cmap="gray")
plt.show()
```

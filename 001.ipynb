# 01.图像的读取（cv2.imread）

```python
import cv2
import os

import matplotlib.pyplot as plt
```

## cv2.imread()
- 1 读取BGR彩色图像
- 0 读取灰度图像
- -1 原图
- 2 8位图片
- 4 任何可能的格式

```python
img_path = "./data/1_DAPI.jpg"

retval = cv2.imread(img_path,flags=1)
```

```python
plt.imshow(cv2.cvtColor(retval,cv2.COLOR_BGR2RGB))
```

## 拓展：从网络获取图像

```python
from urllib import request
import numpy as np

response = request.urlopen("https://profile.csdnimg.cn/8/E/F/0_youcans")
img_net = cv2.imdecode(np.array(bytearray(response.read()),dtype=np.uint8),-1)

plt.imshow(cv2.cvtColor(img_net,cv2.COLOR_BGR2RGB))
```

# 09.图像的裁剪(cv2.selectROI)

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img =  cv2.imread("./data/2_DAPI.jpg")

xmin,ymin,w,h = 180,190,200,200
img_crop = img[xmin:xmin+w,ymin:ymin+h].copy()

plt.subplot(121),plt.axis("off"),plt.title("DAPI")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))

plt.subplot(122),plt.axis("off"),plt.title("CROP")
plt.imshow(cv2.cvtColor(img_crop,cv2.COLOR_BGR2RGB))
```

## 拓展

```python
roi = cv2.selectROI(img,showCrosshair=True,fromCenter=False)
xmin,ymin,w,h = roi
img_roi = img[xmin:xmin+w,ymin:ymin+h].copy()
plt.imshow(cv2.cvtColor(img_roi,cv2.COLOR_BGR2RGB))
```

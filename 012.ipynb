# 012.图像通道的合并

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img_bf = cv2.imread("./data/1_BF.jpg")
# img_dapi = cv2.imread("./data/1_DAPI.jpg")
# img_cy5 = cv2.imread("./data/1_CY5.jpg")
# img_fitc = cv2.imread("./data/1_FITC.jpg")
# img_tritc = cv2.imread("./data/1_TRITC.jpg")
img_bf_b = img_bf[:,:,0]
img_bf_G = img_bf[:,:,1]
img_bf_R = img_bf[:,:,2]

img_merge = cv2.merge([img_bf_b,img_bf_G,img_bf_R])
plt.imshow(img_merge)
```
